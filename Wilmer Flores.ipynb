```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('/Users/treychase/Downloads/Wilmer Flores/wf_batting.csv')
```

```python
fig, ax = plt.subplots(figsize = (10, 6))

ax.plot(df['Year'], df['WAR'], marker = 'o', label = 'WAR')
plt.xlabel('Year')
plt.ylabel('bWAR')

plt.title('Wilmer Flores bWAR 2015-2024')
```

```python
fig, ax = plt.subplots(figsize = (10, 6))

ax.plot(df['Year'], df['RE24'], marker = 'o')
plt.xlabel('Year')
plt.ylabel('RE24 Value')
plt.title('Wilmer Flores RE24 2015-2024')
plt.axhline(0, c='red', linestyle = 'dashed')
#RE24 over the years
```

```python
fig, ax = plt.subplots(figsize = (10, 6))

ax.plot(df['Year'], df['OPS'], marker = 'o')
plt.xlabel('Year')
plt.ylabel('OPS')
plt.title('Wilmer Flores OPS 2015-2024')
```

```python
fig, ax = plt.subplots(figsize = (10, 6))

ax.plot(df['Year'], df['SLG'], marker = 'o', label = 'SLG')
plt.xlabel('Year')


ax.plot(df['Year'], df['OBP'], marker = 'o', label = 'OBP')
plt.xlabel('Year')
plt.ylabel('Percentage')

plt.title('Wilmer Flores OBP and SLG 2015-2024')

plt.legend()
plt.show()
#the slugging and OBP are almost identical in 2024!
```

```python
df['Pull%'] = df['Pull%'].str.replace('%', '')
df['Cent%'] = df['Cent%'].str.replace('%', '')
df['Oppo%'] = df['Oppo%'].str.replace('%', '')
#to change into numbers 
```

```python
df['Pull%'] = pd.to_numeric(df['Pull%'])
df['Cent%'] = pd.to_numeric(df['Cent%'])
df['Oppo%'] = pd.to_numeric(df['Oppo%'])
#for plotting
```

```python
fig, ax = plt.subplots(figsize = (10, 6))

ax.plot(df['Year'], df['Pull%'], marker = 'o', label = 'Pull%')
ax.plot(df['Year'], df['Cent%'], marker = 'o', label = 'Cent%')
ax.plot(df['Year'], df['Oppo%'], marker = 'o', label = 'Oppo%')


plt.xlabel('Year')
plt.ylabel('Percentage')
plt.title('Wilmer Flores Spray 2015-2024')

plt.legend()
plt.show()
```

```python
df['LD%'] = df['LD%'].str.replace('%', '')
df['GB%'] = df['GB%'].str.replace('%', '')
df['FB%'] = df['FB%'].str.replace('%', '')
```

```python
df['LD%'] = pd.to_numeric(df['LD%'])
df['GB%'] = pd.to_numeric(df['GB%'])
df['FB%'] = pd.to_numeric(df['FB%'])
```

```python
fig, ax = plt.subplots(figsize = (10, 6))

ax.plot(df['Year'], df['LD%'], marker = 'o', label = 'Line Drive %')
ax.plot(df['Year'], df['GB%'], marker = 'o', label = 'Ground Ball %')
ax.plot(df['Year'], df['FB%'], marker = 'o', label = 'Fly Ball %')


plt.xlabel('Year')
plt.ylabel('Percentage')
plt.title('Wilmer Flores Contact 2015-2024')

plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (10, 6))

ax.plot(df['Year'], df['EV'], marker = 'o')

plt.xlabel('Year')
plt.ylabel('Exit Velocity')

plt.title('Wilmer Flores Average Exit Velocity 2015-2024')
```
